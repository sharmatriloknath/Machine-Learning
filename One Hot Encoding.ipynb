```python
df = df[df['brand'] != 'Ashok']
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first',sparse=False)
```

```python
X_train_trans = ohe.fit_transform(X_train[['brand','fuel','owner']])
X_test_trans = ohe.transform(X_test[['brand','fuel','owner']])
```

# Number Of Columns Are Hudge
- when we have hudge no of categories that means hudge no of columns in df.
- To avoid this situation we keep top most categories and combine some less frequent categories into a common categories.
- Check the example below.

```python
df.head()
```

```python
df['brand'].value_counts()
```

```python
# combine all the categories into one category with less or equal to 100 counts.
counts = df['brand'].value_counts()
```

```python
df['brand'].replace(counts[counts<=100].index,'others', inplace=True)
```
